```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
DATA_DIR = os.path.abspath("adsp-data/494_bus")
```

```python
# 데이터 불러오기
X = pd.read_excel("dataframe_494_bus_with_index2.xlsx", index_col=0)
X
```

```python
# 파일 이름에서 에러크기 비율과 에러주입 비율 추출
X['error_size_ratio'] = X.index.map(lambda x: float(x.split('_')[-2]))
X['error_injection_ratio'] = X.index.map(lambda x: float(x.split('_')[-1].split('.')[0]))

```

```python
# 첫 번째 열을 제외한 나머지 열들을 사용
counters = X.columns[:-2]
```

```python
# 결과 저장용 리스트 초기화
results_list = []
```

```python
# 모델 정의 함수
def create_model():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dense(11, activation="softmax"),  # 출력 크기 101개
    ])
    optimizer = tf.keras.optimizers.Adam()  # SGD 대신 Adam 사용
    model.compile(loss=tf.keras.losses.CategoricalCrossentropy(), optimizer=optimizer, metrics=["accuracy"])
    return model
```

```python
# 각 counter에 대해 분석
for counter in counters:
    X_col = X[[counter, 'error_size_ratio']]

    # 데이터 분할
    y = X_col.values[:, 1:]  # 에러크기 비율과 에러주입 비율
    X_col = X_col.values[:, 0].reshape(-1, 1)  # counter 값

    # 타겟 값 인코딩
    y_combined = (y * 100).astype(int)  # 에러크기 비율을 인덱스로 사용
    y_combined = tf.one_hot(y_combined, 11)  # 총 11개의 클래스

    # 타겟 값의 불필요한 차원 제거
    y_combined = tf.reshape(y_combined, [-1, 11])

    X_train, X_test, y_train, y_test = train_test_split(X_col, y_combined.numpy(), test_size=0.33)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # 모델 초기화 및 학습
    model = create_model()
    history = model.fit(X_train, y_train, epochs=100, verbose=0)

    # 테스트 데이터 예측
    results = model.predict(X_test)
    results = tf.argmax(results, axis=-1).numpy()
    y_test_labels = tf.argmax(y_test, axis=-1).numpy()

    # 성능 평가
    mse_metric = tf.keras.metrics.MeanSquaredError()
    mse_metric.update_state(y_test_labels, results)
    mse = mse_metric.result().numpy()

    mae_metric = tf.keras.metrics.MeanAbsoluteError()
    mae_metric.update_state(y_test_labels, results)
    mae = mae_metric.result().numpy()

    accuracy_metric = tf.keras.metrics.Accuracy()
    accuracy_metric.update_state(y_test_labels, results)
    accuracy = accuracy_metric.result().numpy()


    # 결과 저장
    results_list.append((counter, mse, mae, accuracy))
```

```python
# Sort results and output top 20 counters (sorted by Accuracy)
results_list.sort(key=lambda x: x[3], reverse=True)  # Accuracy 기준으로 정렬
top_20_counters = results_list[:20]
```

```python
# Result output
for counter, mse, mae, accuracy in top_20_counters:
    print(f"Counter: {counter}, MSE: {mse}, MAE: {mae}, Accuracy: {accuracy}")
```

```python
# 시각화: 상위 20개 counter에 대해 실제값과 예측값 비교
for counter, _, _, _ in top_20_counters:
    X_col = X[[counter, 'error_size_ratio']]

    # 데이터 분할
    y = X_col.values[:, 1]  # 에러크기 비율
    X_col = X_col.values[:, 0].reshape(-1, 1)  # counter 값

    # 타겟 값 인코딩
    y_combined = (y * 100).astype(int)  # 에러크기 비율을 인덱스로 사용
    y_combined = tf.one_hot(y_combined, 11)  # 총 11개의 클래스

    # 타겟 값의 불필요한 차원 제거
    y_combined = tf.reshape(y_combined, [-1, 11])

    X_train, X_test, y_train, y_test = train_test_split(X_col, y_combined.numpy(), test_size=0.33)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # 모델 초기화 및 학습
    model = create_model()
    model.fit(X_train, y_train, epochs=100, verbose=0)

    # 테스트 데이터 예측
    results = model.predict(X_test)
    results = tf.argmax(results, axis=-1).numpy()
    y_test_labels = tf.argmax(y_test, axis=-1).numpy()

    # 시각화
    f = plt.figure()
    plt.plot(list(range(len(y_test_labels))), y_test_labels, color="red", label="True Labels")
    plt.plot(list(range(len(y_test_labels))), results, color="blue", label="Predicted Labels")
    plt.title(f"Counter: {counter}")
    plt.ylabel("Error rate")
    f.set_figwidth(15)
    plt.legend()
    plt.show()
```
